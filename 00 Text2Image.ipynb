### Imports

```python
from diffusers import AutoPipelineForText2Image
import torch
```

### Load Model

```python
pipeline = AutoPipelineForText2Image.from_pretrained(
	"runwayml/stable-diffusion-v1-5", 
    torch_dtype=torch.float16, 
    variant="fp16"
).to("cuda")
generator = torch.Generator("cuda").manual_seed(31)
```

### Generate Image 1: Default Sample

```python
image = pipeline("Astronaut in a jungle, cold color palette, muted colors, detailed, 8k", generator=generator).images[0]
image
```

### Generate Image 2: Own Sample

```python
image = pipeline("Side view of a black Mercedes", generator=generator).images[0]
image
```

### Generate Image 3: Breaking Down Seprate Steps

```python
images = pipeline("close up side view of a car with a passenger", generator=generator).images
```
